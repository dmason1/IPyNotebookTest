```python
# ds is a list of 'databoxes', each containing a 0.5s long time sample.
ds=s.data.load_multiple(moredpaths[:25],quiet=True)

# this splits apart the different columns of each databox.
# the columns are:
#     0 : time axis
#     3 : real part of signal
#     4 : imaginary part of signal
# ( x(t) has been mixed with a complex exponential at 788 kHz, so that the mechanical signal
#   x(t) appears approximately at 0Hz.  Real/Imag parts of this signal give us two quadratures of the motion)
rawts = [d[0] for d in ds]
rawxs = [d[3] for d in ds]
rawys = [d[4] for d in ds]
rawzs = [d[3]+1j*d[4] for d in ds]
rawthetas  = [np.arctan2(y,x) for x,y in zip(rawxs,rawys)]
# amplitudes of the signals
rawrs = [abs(x+1j*y) for x,y in zip(rawxs,rawys)]
# amplitude^2 of the signals...
rawr2s = [abs(x+1j*y)**2 for x,y in zip(rawxs,rawys)]
# x^2, y^2...
rawx2s = [x**2 for x in rawxs]
rawy2s = [y**2 for y in rawys]

# Okay, this was something I started doing at some point...
# "raw" means the original data sample
# The variables below have been 'coarsened', meaning that neighboring points have been binned together
# (The original time samples had 57000 samples/second, but we don't really care about stuff that fast.
# For some purposes, I'd rather have a smaller array without all the excess high-frequency stuff)
ts = [s.fun.coarsen_array(t,20) for t in rawts]
xs = [s.fun.coarsen_array(x,20) for x in rawxs]
ys = [s.fun.coarsen_array(y,20) for y in rawys]
zs = [s.fun.coarsen_array(z,20) for z in rawzs]
thetas = [s.fun.coarsen_array(th,20) for th in rawthetas]
rs = [s.fun.coarsen_array(r,20) for r in [abs(x+1j*y) for x,y in zip([d[3] for d in ds],[d[4] for d in ds])]]
r2s = [s.fun.coarsen_array(r,20)**2 for r in [abs(x+1j*y) for x,y in zip([d[3] for d in ds],[d[4] for d in ds])]]


# things like tt, xx, yy, etc. are arrays where I stitch all of the individual time samples together,
# making them into one big long sample.
tt = array([ts[i]+i*ts[0][-1] for i in range(len(ts))]).flatten() 
rawtt = array([rawts[i]+i*rawts[0][-1] for i in range(len(rawts))]).flatten() 

xx = array(xs).flatten()
rawxx = array(rawxs).flatten()

yy = array(ys).flatten()
rawyy = array(rawys).flatten()

zz = array(zs).flatten()
rawzz = array(rawzs).flatten()

rawthth = array(rawthetas).flatten()
thth = array(thetas).flatten()

rawxx2 = array(rawx2s).flatten()

rr = array(rs).flatten()
rawrr = array(rawrs).flatten()

rr2 = array(r2s).flatten()
rawrr2 = array(rawr2s).flatten()
```

```python
# This is where I compute histograms of the individual time samples
xshist = [np.histogram(x,bins=30) for x in xs]
yshist = [np.histogram(y,bins=30) for y in ys]

# computing standard deviations of individual time samples
xsvar = [np.std(x) for x in xs]
ysvar = [np.std(y) for y in ys]

# histograms and standard deviations of the long time sample.
xxhist = np.histogram(xx,bins=60)
yyhist = np.histogram(yy,bins=60)

rrhist = np.histogram(rr,bins=60)
rawrrhist = np.histogram(rawrr,bins=60)

rr2hist = np.histogram(rr2,bins=60)
rawrr2hist = np.histogram(rawrr2,bins=60)

rawxx2hist = np.histogram(rawxx2,bins=60)

xxvar = np.std(xx)
yyvar = np.std(yy)
```
